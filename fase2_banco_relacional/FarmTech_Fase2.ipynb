```python
import pandas as pd
import sqlite3
import os
```

```python
# Ler a base da Fase 1 (com meteo)
df_talhoes = pd.read_csv("base_fase1_com_meteo.csv")
df_talhoes.head()
```

```python
# Cria (ou abre) o banco SQLite
conn = sqlite3.connect("farmtech.db")

# Só pra conferir que conectou
conn
```

```python
# Grava o DataFrame em uma tabela chamada "talhoes"
df_talhoes.to_sql("talhoes", conn, if_exists="replace", index=False)

# Verificar se deu certo: fazer um SELECT
cursor = conn.cursor()
cursor.execute("SELECT * FROM talhoes LIMIT 5;")
linhas = cursor.fetchall()

linhas
```

```python
# Criar tabela para leituras de sensores (para ser usada na Fase 3)
cursor.execute("""
CREATE TABLE IF NOT EXISTS leituras_sensores (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    talhao_id INTEGER,
    data_hora TEXT,
    umidade REAL,
    ph REAL,
    nutrientes REAL,
    origem TEXT, -- 'simulado' ou 'real'
    FOREIGN KEY (talhao_id) REFERENCES talhoes(talhao_id)
);
""")

conn.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
conn.close()
print("Banco farmtech.db criado e conexão fechada.")
```
